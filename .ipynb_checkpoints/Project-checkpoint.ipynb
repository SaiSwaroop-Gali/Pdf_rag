### Extracting atomic elements from PDF using unstructured

````python
def partition_document(file_path: str):
    
    print(f"📄 Partitioning document: {file_path}")
    
    elements = partition_pdf(
        filename=file_path,  # Path of PDF file
        strategy="hi_res", # Use the most accurate (but slower) processing method of extraction
        infer_table_structure=True, # Keep tables as HTML
        extract_image_block_types=["Image"], # Grab images found in the PDF
        extract_image_block_to_payload=True # Store images as base64 data you can actually use
    )
    
    print(f"✅ Extracted {len(elements)} elements")
    return elements



from pathlib import Path

pdf_folder = Path("./data")

all_elements = []

for pdf_file in pdf_folder.glob("*.pdf"):
    elements = partition_document(str(pdf_file))
    all_elements.extend(elements)

print(f"\n✅ Total elements extracted: {len(all_elements)}")

# Test with your PDF file
#file_path = "./data/test.pdf"  # PDF path
#elements = partition_document(file_path)
````

### Creating intelligent chunks using title-based strategy

````python
def create_chunks_by_title(all_elements):
    
    print("🔨 Creating smart chunks...")
    
    chunks = chunk_by_title(
        all_elements, # The parsed PDF elements
        max_characters=3000, # Hard limit - never exceed 3000 characters per chunk
        new_after_n_chars=2400, # Try to start a new chunk after 2400 characters
        combine_text_under_n_chars=500 # Merge tiny chunks under 500 chars with neighbors
    )
    
    print(f"✅ Created {len(chunks)} chunks")
    return chunks

# Create chunks
chunks = create_chunks_by_title(all_elements)
````

````python
def separate_content_types(chunk):
    """Analyze what types of content are in a chunk"""
    content_data = {
        'text': chunk.text,
        'tables': [],
        'images': [],
        'types': ['text']
    }
    
    # Check for tables and images in original elements
    if hasattr(chunk, 'metadata') and hasattr(chunk.metadata, 'orig_elements'):
        for element in chunk.metadata.orig_elements:
            element_type = type(element).__name__
            
            # Handle tables
            if element_type == 'Table':
                content_data['types'].append('table')
                table_html = getattr(element.metadata, 'text_as_html', element.text)
                content_data['tables'].append(table_html)
            
            # Handle images
            elif element_type == 'Image':
                if hasattr(element, 'metadata') and hasattr(element.metadata, 'image_base64'):
                    content_data['types'].append('image')
                    content_data['images'].append(element.metadata.image_base64)
    
    content_data['types'] = list(set(content_data['types']))
    return content_data

def create_ai_enhanced_summary(text: str, tables: List[str], images: List[str]) -> str:
    """Create AI-enhanced summary for mixed content"""
    
    try:
        # Build the text prompt
        prompt_text = f"""You are creating a searchable description for document content retrieval.

        CONTENT TO ANALYZE:
        TEXT CONTENT:
        {text}

        """
        
        # Add tables if present
        if tables:
            prompt_text += "TABLES:\n"
            for i, table in enumerate(tables):
                prompt_text += f"Table {i+1}:\n{table}\n\n"
        
                prompt_text += """
                YOUR TASK:
                Generate a comprehensive, searchable description that covers:

                1. Key facts, numbers, and data points from text and tables
                2. Main topics and concepts discussed  
                3. Questions this content could answer
                4. Visual content analysis (charts, diagrams, patterns in images)
                5. Alternative search terms users might use

                Make it detailed and searchable - prioritize findability over brevity.

                SEARCHABLE DESCRIPTION:"""

        # Build message content starting with text
        message_content = [{"type": "text", "text": prompt_text}]
        
        # Add images to the message
        for image_base64 in images:
            message_content.append({
                "type": "image_url",
                "image_url": {"url": f"data:image/jpeg;base64,{image_base64}"}
            })
        
        # Send to AI and get response
        message = HumanMessage(content=message_content)
        response = llm.invoke([message])
        
        return response.content
        
    except Exception as e:
        print(f"     ❌ AI summary failed: {e}")
        # Fallback to simple summary
        summary = f"{text[:300]}..."
        if tables:
            summary += f" [Contains {len(tables)} table(s)]"
        if images:
            summary += f" [Contains {len(images)} image(s)]"
        return summary

def summarise_chunks(chunks):
    """Process all chunks with AI Summaries"""
    print("🧠 Processing chunks with AI Summaries...")
    
    langchain_documents = []
    total_chunks = len(chunks)
    
    for i, chunk in enumerate(chunks):
        current_chunk = i + 1
        print(f"   Processing chunk {current_chunk}/{total_chunks}")
        
        # Analyze chunk content
        content_data = separate_content_types(chunk)
        
        # Debug prints
        print(f"     Types found: {content_data['types']}")
        print(f"     Tables: {len(content_data['tables'])}, Images: {len(content_data['images'])}")
        
        # Create AI-enhanced summary if chunk has tables/images
        if content_data['tables'] or content_data['images']:
            print(f"     → Creating AI summary for mixed content...")
            try:
                enhanced_content = create_ai_enhanced_summary(
                    content_data['text'],
                    content_data['tables'], 
                    content_data['images']
                )
                print(f"     → AI summary created successfully")
                print(f"     → Enhanced content preview: {enhanced_content[:200]}...")
            except Exception as e:
                print(f"     ❌ AI summary failed: {e}")
                enhanced_content = content_data['text']
        else:
            print(f"     → Using raw text (no tables/images)")
            enhanced_content = content_data['text']
        
        # Create LangChain Document with rich metadata
        doc = Document(
            page_content=enhanced_content,
            metadata={
                "original_content": json.dumps({
                    "raw_text": content_data['text'],
                    "tables_html": content_data['tables'],
                    "images_base64": content_data['images']
                })
            }
        )
        
        langchain_documents.append(doc)
    
    print(f"✅ Processed {len(langchain_documents)} chunks")
    return langchain_documents


# Process chunks with AI
processed_chunks = summarise_chunks(chunks)
````

### Export processed chunks to clean JSON format

````python
def export_chunks_to_json(chunks, filename="chunks_export.json"):
    
    export_data = []
    
    for i, doc in enumerate(chunks):
        chunk_data = {
            "chunk_id": i + 1,
            "enhanced_content": doc.page_content,
            "metadata": {
                "original_content": json.loads(doc.metadata.get("original_content", "{}"))
            }
        }
        export_data.append(chunk_data)
    
    # Save to file
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(export_data, f, indent=2, ensure_ascii=False)
    
    print(f"✅ Exported {len(export_data)} chunks to {filename}")
    return export_data

# Export your chunks
json_data = export_chunks_to_json(processed_chunks)
````

### Creating ChromaDB vector store data injestion through openai text embedding 3 small (openai api key)

````python
import time
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma

def create_vector_store(documents, persist_directory="dbv1/chroma_db"):
    print("🔮 Creating embeddings and storing in ChromaDB with rate-limiting...")
    
    # Initialize the embedding model with your optimized 512 dimensions
    embedding_model = OpenAIEmbeddings(
        model="text-embedding-3-small",
        
    )
    
    # 1. Define a safe batch size based on your 40,000 TPM limit.
    # 40 chunks * ~750 tokens = ~30,000 tokens per batch (safely below 40k).
    batch_size = 40 
    
    # 2. Initialize the Chroma database with just the FIRST batch
    print("--- Creating vector store with initial batch ---")
    first_batch = documents[:batch_size]
    
    vectorstore = Chroma.from_documents(
        documents=first_batch,
        embedding=embedding_model,
        persist_directory=persist_directory, 
        collection_metadata={"hnsw:space": "cosine"}
    )
    
    # 3. Loop through and add the rest of the chunks incrementally
    total_chunks = len(documents)
    for i in range(batch_size, total_chunks, batch_size):
        # Calculate the next slice of chunks
        current_batch = documents[i:i + batch_size]
        
        # Calculate human-readable progress stats
        progress_start = i
        progress_end = min(i + batch_size, total_chunks)
        
        # Enforce the time constraint: wait 60 seconds to reset OpenAI's TPM tracker
        print(f"⏳ Rate limit protection: Waiting 60 seconds before processing next batch...")
        time.sleep(60)
        
        print(f"📦 Indexing chunks {progress_start} to {progress_end} of {total_chunks}...")
        vectorstore.add_documents(documents=current_batch)
        
    print("--- Finished creating vector store ---")
    print(f"✅ Vector store created and saved to {persist_directory}")
    return vectorstore

# Create the vector store safely
db = create_vector_store(processed_chunks)
````

### Triples Function

````python
import json
import re
from langchain_core.messages import HumanMessage

def extract_triples(text):

    prompt = f"""
Extract ONLY AI-related entities and relationships.

Focus on:
- Methods
- Frameworks
- Models
- Datasets
- Benchmarks
- Metrics
- Concepts

Ignore:
- Authors
- Universities
- Affiliations
- References

Return ONLY a JSON array.

Format:

[
  {{
    "source": "entity",
    "relationship": "RELATIONSHIP",
    "target": "entity"
  }}
]

Text:
{text}
"""

    response = kg_llm.invoke(
        [HumanMessage(content=prompt)]
    )

    content = response.content.strip()

    # Remove markdown fences
    content = re.sub(r"^```json\s*", "", content)
    content = re.sub(r"^```\s*", "", content)
    content = re.sub(r"\s*```$", "", content)

    try:
        return json.loads(content)

    except Exception:
        print("Failed response:")
        print(content)
        return []
````

````python
def graph_retrieve(entity_name):

    query = """
    MATCH (s)-[r]->(t)
    WHERE toLower(s.name) CONTAINS toLower($entity)
    RETURN
        s.name AS source,
        type(r) AS relationship,
        t.name AS target
    LIMIT 20
    """

    return graph.query(
        query,
        {"entity": entity_name}
    )
````

### Cleaning data filling the target values

````python
repaired_triples = []

for t in all_triples:

    # already valid
    if (
        isinstance(t, dict)
        and "source" in t
        and "relationship" in t
        and "target" in t
    ):
        repaired_triples.append(t)

    # missing target
    elif (
        isinstance(t, dict)
        and "source" in t
        and "relationship" in t
    ):

        rel = t["relationship"]

        if rel.startswith("is_"):

            target = rel.replace("is_", "")
            target = target.replace("_", " ").title()

            repaired_triples.append(
                {
                    "source": t["source"],
                    "relationship": "IS_A",
                    "target": target
                }
            )

        else:

            print("Could not repair:")
            print(t)

print(
    f"Repaired triples count: {len(repaired_triples)}"
)
````

### The triples which has all 3 key value pairs

````python
clean_triples = []

for t in all_triples:

    if (
        isinstance(t, dict)
        and "source" in t
        and "relationship" in t
        and "target" in t
    ):
        clean_triples.append(t)

print(f"Original triples: {len(all_triples)}")
print(f"Valid triples: {len(clean_triples)}")
print(f"Repaired triples count: {len(repaired_triples)}")
````

### Finding list type entries

````python
for i, t in enumerate(repaired_triples):

    try:
        hash(
            (
                t["source"],
                t["relationship"],
                t["target"]
            )
        )

    except Exception:
        print(f"\nProblem triple #{i}")
        print(t)
````

### Solving list type keyvalues using edge expansion method

````python
expanded_triples = []

for t in repaired_triples:

    source = str(t["source"])
    relationship = str(t["relationship"])
    target = t["target"]

    if isinstance(target, list):

        for item in target:

            expanded_triples.append(
                {
                    "source": source,
                    "relationship": relationship,
                    "target": str(item)
                }
            )

    else:

        expanded_triples.append(
            {
                "source": source,
                "relationship": relationship,
                "target": str(target)
            }
        )

print(
    f"Expanded triples: {len(expanded_triples)}"
)
````

### Removing duplicates

````python
unique_triples = {
    (
        t["source"].strip(),
        t["relationship"].strip(),
        t["target"].strip()
    )
    for t in expanded_triples
}

print(f"Unique triples: {len(unique_triples)}")
````

## Vector Retriever

````python
import json
from langchain_core.messages import HumanMessage

def generate_final_answer(query):

    chunks = vector_retriever.invoke(query)

    if not chunks:

        return (
            "Insufficient information available "
            "in the local PDF knowledge base."
        )

    prompt_text = f"""
Answer the question ONLY from the retrieved content.

QUESTION:
{query}

CONTENT:
"""

    image_count = 0
    table_count = 0

    message_content = []

    for i, chunk in enumerate(chunks):

        prompt_text += (
            f"\n\n========== DOCUMENT {i+1} ==========\n"
        )

        prompt_text += (
            f"\nRETRIEVED SUMMARY:\n"
            f"{chunk.page_content}\n"
        )

        if "original_content" not in chunk.metadata:
            continue

        data = json.loads(
            chunk.metadata["original_content"]
        )

        # Raw Text
        raw_text = data.get(
            "raw_text",
            ""
        )

        if raw_text:

            prompt_text += (
                f"\nRAW TEXT:\n"
                f"{raw_text}\n"
            )

        # Tables
        tables = data.get(
            "tables_html",
            []
        )

        if tables:

            prompt_text += "\nTABLES:\n"

            for table in tables:

                table_count += 1

                prompt_text += (
                    f"\nTABLE {table_count}:\n"
                    f"{table}\n"
                )

    prompt_text += """

Use all available information.

If the answer is not present in the retrieved content,
respond exactly:

INSUFFICIENT_INFORMATION
"""

    message_content.append(
        {
            "type": "text",
            "text": prompt_text
        }
    )

    # Add Images
    for chunk in chunks:

        if "original_content" not in chunk.metadata:
            continue

        data = json.loads(
            chunk.metadata["original_content"]
        )

        images = data.get(
            "images_base64",
            []
        )

        for image in images:

            image_count += 1

            message_content.append(
                {
                    "type": "image_url",
                    "image_url": {
                        "url":
                        f"data:image/jpeg;base64,{image}"
                    }
                }
            )

    print(f"Tables sent: {table_count}")
    print(f"Images sent: {image_count}")

    response = rag_llm.invoke(
        [HumanMessage(content=message_content)]
    )

    answer = response.content.strip()

    if answer == "INSUFFICIENT_INFORMATION":

        return (
            "Insufficient information available "
            "in the local PDF knowledge base."
        )

    return answer
````

## Dynamic entities / Entity exctraction agent

````python
from langchain_openai import ChatOpenAI

ee_llm = ChatOpenAI(
    model="gpt-4.1-mini",
    temperature=0
)
````

````python
# from langchain_google_genai import ChatGoogleGenerativeAI
# from langchain_core.messages import HumanMessage
# import json

# entity_llm = ChatGoogleGenerativeAI(
#     model="gemini-2.5-flash",
#     temperature=0
# )
````

````python
def entity_agent(question):

    prompt = f"""
You are an entity extraction agent.

Extract important AI entities, frameworks, methods,
datasets, benchmarks, concepts and systems.

Return ONLY a JSON list.

Examples:

Question:
How does Reflexion use episodic memory?

Output:
["Reflexion", "episodic memory"]

Question:
Compare ReAct and Reflexion.

Output:
["ReAct", "Reflexion"]

Question:
How does SELF-RAG differ from RAG?

Output:
["SELF-RAG", "RAG"]

Question:
{question}
"""

    response = ee_llm.invoke(
        [HumanMessage(content=prompt)]
    )

    content = response.content.strip()

    content = content.replace("```json", "")
    content = content.replace("```", "")

    try:
        return json.loads(content)

    except Exception:
        print(content)
        return []
````

````python
entity_agent(
    "How does Reflexion use episodic memory?"
)
````

````python
entity_agent(
    "Compare ReAct and Reflexion"
)
````

### Connecting agent to neo4j

````python
def graph_retrieve(entity, limit=20):

    query = """
    MATCH (a)-[r]->(b)
    WHERE toLower(a.name)=toLower($entity)

    RETURN
        a.name AS source,
        type(r) AS relationship,
        b.name AS target

    LIMIT $limit
    """

    return graph.query(
        query,
        {
            "entity": entity,
            "limit": limit
        }
    )
````

### Graph retriever

````python
def retrieve_entities(question):

    entities = entity_agent(question)

    all_results = []

    for entity in entities:

        print(f"Searching: {entity}")

        triples = graph_retrieve(entity)

        all_results.extend(triples)

    return all_results
````

````python
results = retrieve_entities(
    "How does Reflexion use episodic memory?"
)

print(len(results))
print(results)
````

````python
def graph_retrieve(entity, limit=20):

    query = """
    MATCH (a)-[r]->(b)

    WHERE
        toLower(a.name) CONTAINS toLower($entity)
        OR
        toLower(b.name) CONTAINS toLower($entity)

    RETURN
        a.name AS source,
        type(r) AS relationship,
        b.name AS target

    LIMIT $limit
    """

    return graph.query(
        query,
        {
            "entity": entity,
            "limit": limit
        }
    )
````

````python
def build_graph_context(question):

    triples = retrieve_entities(question)

    unique_lines = set()

    for t in triples:

        line = (
            f"{t['source']} "
            f"{t['relationship']} "
            f"{t['target']}"
        )

        unique_lines.add(line)

    return "\n".join(sorted(unique_lines))
````

````python
question = "How does Reflexion use episodic memory?"

graph_context = build_graph_context(question)

print(graph_context)
````

### Hybrid Retriever

````python
import json

def hybrid_retrieve(question):

    # =========================
    # VECTOR RETRIEVAL
    # =========================

    docs = vector_retriever.invoke(question)

    vector_context = ""

    image_data = []

    for i, doc in enumerate(docs[:3]):

        vector_context += (
            f"\n\n========== DOCUMENT {i+1} ==========\n"
        )

        # Retrieved summary / searchable content
        vector_context += (
            f"\nSUMMARY:\n"
            f"{doc.page_content}\n"
        )

        # Original extracted content
        if "original_content" in doc.metadata:

            try:

                original_data = json.loads(
                    doc.metadata["original_content"]
                )

                # Raw Text
                raw_text = original_data.get(
                    "raw_text",
                    ""
                )

                if raw_text:

                    vector_context += (
                        f"\nRAW TEXT:\n"
                        f"{raw_text}\n"
                    )

                # Tables
                tables_html = original_data.get(
                    "tables_html",
                    []
                )

                if tables_html:

                    vector_context += "\nTABLES:\n"

                    for table_num, table in enumerate(
                        tables_html,
                        start=1
                    ):

                        vector_context += (
                            f"\nTABLE {table_num}:\n"
                            f"{table}\n"
                        )

                # Images
                images_base64 = original_data.get(
                    "images_base64",
                    []
                )

                image_data.extend(images_base64)

            except Exception as e:

                print(
                    f"Error parsing original_content: {e}"
                )

    # =========================
    # GRAPH RETRIEVAL
    # =========================

    entities = entity_agent(question)

    graph_lines = set()

    for entity in entities:

        triples = graph_retrieve(
            entity,
            limit=10
        )

        for triple in triples:

            graph_lines.add(
                f"{triple['source']} "
                f"{triple['relationship']} "
                f"{triple['target']}"
            )

    graph_context = "\n".join(
        sorted(graph_lines)
    )

    # =========================
    # RETURN
    # =========================

    return {
        "entities": entities,
        "vector_context": vector_context,
        "graph_context": graph_context,
        "images": image_data,
        "retrieved_docs": docs
    }
````

````python
result = hybrid_retrieve(
    "How does Reflexion use episodic memory?"
)

print("ENTITIES")
print(result["entities"])

print("\nGRAPH")
print(result["graph_context"][:1000])

print("\nVECTOR")
print(result["vector_context"][:1000])
````

````python
def answer_question(question):

    context = hybrid_retrieve(question)

    prompt = f"""
You are an AI research assistant.

Use ONLY the supplied context.

Example 1:

Question:
How does Reflexion use episodic memory?

Context:
Reflexion stores self-reflection outputs in memory.

Answer:
Reflexion uses episodic memory by storing self-reflection outputs and past experiences.

Example 2:

Question:
Who won the FIFA World Cup 2026?

Context:
Reflexion
ReAct
SELF-RAG

Answer:
INSUFFICIENT_INFORMATION

------------------------------------------------

GRAPH CONTEXT:
{context['graph_context']}

VECTOR CONTEXT:
{context['vector_context']}

QUESTION:
{question}

If the answer is not explicitly supported by the context,
return ONLY:

INSUFFICIENT_INFORMATION

ANSWER:
"""

    response = llm.invoke(prompt)

    answer = response.content.strip()

    if answer == "INSUFFICIENT_INFORMATION":

        return (
            "Data related to this question is not "
            "available in the local knowledge database."
        )

    return answer
````

### Telemetrics & Evals

````python
import json
from datetime import datetime

TELEMETRY_FILE = "telemetry.jsonl"

def log_telemetry(data):

    with open(
        TELEMETRY_FILE,
        "a",
        encoding="utf-8"
    ) as f:

        f.write(
            json.dumps(
                data,
                ensure_ascii=False
            )
            + "\n"
        )
````

````python
from datetime import datetime

def answer_with_telemetry(question):

    context = hybrid_retrieve(question)

    answer = answer_question(question)

    telemetry = {

        "timestamp":
            datetime.now().isoformat(),

        "question":
            question,

        "entities":
            context["entities"],

        "graph_triples":
            len(
                context["graph_context"]
                .split("\n")
            ),

        "retrieved_docs":
            len(
                context["retrieved_docs"]
            ),

        "vector_context_length":
            len(
                context["vector_context"]
            ),

        "answer":
            answer
    }

    log_telemetry(
        telemetry
    )

    return answer
````

````python
answer_with_telemetry(
    "How does Reflexion use episodic memory?"
)
````
